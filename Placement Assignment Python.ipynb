### Question 10 - Write a program to count the number of verbs, nouns, pronouns, and adjectives in a given particular phrase or paragraph, and return their respective count as a dictionary.

Note -
1. Write code comments wherever required for code
2. You have to write at least 2 additional test cases in which your program will run successfully and provide an explanation for the same.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize, sent_tokenize
stop_words = set(stopwords.words('english'))


def getDic(txt):
    dic ={}
    nouns = 0
    pronouns = 0
    verbs =0 
    adjectives = 0
    tokenized = sent_tokenize(txt)
    for i in tokenized:

        # Word tokenizers is used to find the words
        # and punctuation in a string
        wordsList = nltk.word_tokenize(i)

        # removing stop words from wordList
        wordsList = [w for w in wordsList if not w in stop_words]

        # remove the call to nltk.pos_tag if `sentence` is a list of tuples as described above
        for word, pos in nltk.pos_tag(wordsList): 
            if pos in ['NN', "NNP", "NNS", "NNPS"]: 
                nouns = nouns + 1

            if pos in ['PRP', "PRP$", "NNS"]: 
                pronouns = pronouns + 1

            if pos in ['VB', "VBD", "VBG", 'VBN', "VBP", "VBZ"]: 
                verbs = verbs + 1

            if pos in ['JJ', "JJR", "JJS"]: 
                adjectives = adjectives + 1

    dic ={
        "nouns":nouns,
        "pronouns":pronouns,
        "verbs":verbs,
        "adjectives":adjectives
    }

    print(dic)
```
